# ðŸ”´ Task 15: Data Wrangling: Join, Combine, and Reshape.

```python
import numpy as np
import pandas as pd
from IPython.display import display
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 10)
pd.set_option('display.width', 1000)
```

```python
import_dataFrame = pd.read_csv('1.csv')
import_dataFrame2 = pd.read_csv('2.csv')
```

## 1. Merge two DataFrames on a single key.

```python
merger = pd.merge(import_dataFrame, import_dataFrame2, on='Index')
display(merger)
```

## 2. Merge two DataFrames on multiple keys.

```python
merger2 = pd.merge(import_dataFrame, import_dataFrame2, on=['Index', 'First Name'])
display(merger2)
```

## 3. Perform an outer join, inner join, left join, and right join.

```python
merger_Outer = pd.merge(import_dataFrame, import_dataFrame2, on=['Index', 'First Name'], how = 'outer')
merger_Inner = pd.merge(import_dataFrame, import_dataFrame2, on=['Index', 'First Name'], how = 'inner')
merger_Left = pd.merge(import_dataFrame, import_dataFrame2, on=['Index', 'First Name'], how = 'left')
merger_Right = pd.merge(import_dataFrame, import_dataFrame2, on=['Index', 'First Name'], how = 'right')
display(merger_Outer)
display(merger_Inner)
display(merger_Left)
display(merger_Right)
```

## 4. Concatenate two DataFrames along rows.

```python
concatenate = pd.concat([import_dataFrame, import_dataFrame2], ignore_index=True)
display(concatenate)
```

## 5. Concatenate two DataFrames along columns.

```python
concatenate2 = pd.concat([import_dataFrame, import_dataFrame2], ignore_index=True, axis = 1)
display(concatenate2)
```

## 6. Concatenate a list of DataFrames.

```python
import_dataFrame3 = pd.read_csv("3.csv")
listed = [import_dataFrame, import_dataFrame2, import_dataFrame3]
concat = pd.concat(listed, ignore_index=True)
display(concat)
```

## 7. Reshape data using the melt function to go from wide to long format.

```python
melted = import_dataFrame.melt(
    id_vars=['Index', 'Customer Id', 'First Name', 'Last Name', 'Company', 'Subscription Date', 'Website'],
    var_name='Contact Info',
    value_name='Value'
)
display(melted)
```

## 8. Create a pivot table to summarize data.

```python
pivot = melted.pivot_table(
    index=['Index', 'Customer Id', 'First Name', 'Last Name', 'Company', 'Subscription Date', 'Website'],
    columns='Contact Info',
    values='Value',
    aggfunc='sum'
)
display(pivot)
```

## 9. Group data by one or more columns and perform aggregation functions (e.g., sum, mean, count).

```python
import_dataFrame['Subscription Date'] = pd.to_datetime(import_dataFrame['Subscription Date'])

customer_counts = import_dataFrame.groupby(['First Name', 'Last Name']).size().reset_index(name='Count')
display(customer_counts)

avg_subscription = import_dataFrame.groupby('Company')['Subscription Date'].mean().reset_index(name='Avg Subscription Date')
display(avg_subscription)

phone_count = import_dataFrame.groupby('Company')[['Phone 1', 'Phone 2']].count().reset_index()
display(phone_count)

company_emails = import_dataFrame.groupby('Company')['Email'].apply(lambda x: ', '.join(x.unique())).reset_index(name='Emails')
display(company_emails)
```

## 10. Apply multiple aggregation functions to grouped data.

```python
grouped = import_dataFrame.groupby('Company')

company_stats = grouped.agg({
    'Subscription Date': ['min', 'max', 'mean'],
    'Customer Id': 'count',
    'First Name': 'nunique',
    'Last Name': 'nunique'
})

company_stats.columns = ['Min Subscription', 'Max Subscription', 'Avg Subscription',
                        'Total Customers', 'Unique First Names', 'Unique Last Names']

display(company_stats)
```

## 11. Use the groupby function to group data and apply custom functions.

```python
import_dataFrame['Subscription Date'] = pd.to_datetime(import_dataFrame['Subscription Date'])

grouped = import_dataFrame.groupby('Company', group_keys=False)

def custom_agg(group):
    return pd.Series({
        'Customer Count': group['Customer Id'].count(),
        'Earliest Subscription': group['Subscription Date'].min()
    })

result = grouped.apply(custom_agg, include_groups=False).reset_index()

display(result)
```

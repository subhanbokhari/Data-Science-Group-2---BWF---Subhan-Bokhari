## Import Necessary Libraries

```python
import pandas as pd
import re
import spacy
nlp = spacy.load('en_core_web_sm')
```

## Using Spacy as nltk was showing some errors that I tried my best to eradicate but it was a issue that I couldn't resolve.

## Import DataSet

```python
df_imp = pd.read_csv('titanic.csv')
```

```python
display(df_imp)
```

## Convert all of the Data to Lower case to make it easier to make it recognized

```python
df_imp = df_imp.map(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
display(df_imp)
```

## Check and remove if any at all link is used in the data

```python
url_pattern = re.compile(r'https?://\S+')
def remove_urls(text):
    return url_pattern.sub('', text)
df_imp['Name'] = df_imp['Name'].apply(remove_urls)
```

## Check and remove any irregular Symbol or characcter used

```python
df_imp = df_imp.replace(to_replace=r'[^\w\s]', value='', regex=True)
```

```python
display(df_imp)
```

## Remove if any digit found

```python
df_imp['Ticket'] = df_imp['Ticket'].replace(to_replace=r'\d', value='', regex=True)
```

```python
display(df_imp)
```

## Tokenization of Names using Spacy

```python
def tokenize_name(name):
    return [token.text for token in nlp(name)]
df_imp['Name_tokens'] = df_imp['Name'].apply(lambda x: tokenize_name(str(x)))
print(df_imp[['Name', 'Name_tokens']].head())
```

## Remove any Stopwords if they are present

```python
def remove_stopwords(name):
    doc = nlp(name)
    return [token.text for token in doc if not token.is_stop]

df_imp['Name_cleaned'] = df_imp['Name'].apply(lambda x: remove_stopwords(str(x)))
print(df_imp[['Name', 'Name_cleaned']].head())
```

## Check and Replace any word that is misspelled

## spaCy allows only lemmatization and it is okay as it is more efficient than stemming

```python
def lemmatize_text(text):
    doc = nlp(text)
    return [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
df_imp['lemmatized_messages'] = df_imp['Name'].apply(lemmatize_text)
print(df_imp[['Name', 'lemmatized_messages']].head())
```

#### In this code, we outlined the primary preprocessing steps for constructing an NLP model, which involve cleaning the text, tokenizing it, removing stopwords, and applying stemming or lemmatization. These steps are crucial for enhancing model performance by minimizing text noise and transforming the text into a more structured format suitable for analysis by the model.

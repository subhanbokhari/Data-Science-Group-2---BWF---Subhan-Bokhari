## Import the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import seaborn as sns
```

## Generate a sample dataset with 4 clusters

```python
X, y = make_blobs(n_samples=500, centers=4, n_features=2, random_state=42)
```

## Use Elbow method to get the # of clusters

```python
emc = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(X)
    emc.append(kmeans.inertia_)
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), emc)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCS')
plt.show()
```

## K-means clustering on the # of clusters found

```python
kms = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kms = kmeans.fit_predict(X)
```

## Visual Representation of the Clusters

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_kms, palette='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=100, c='yellow', label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

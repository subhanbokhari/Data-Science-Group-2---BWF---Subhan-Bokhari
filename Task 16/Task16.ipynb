## Basic Plots:

#### 1. Create a simple line plot.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [4,7,9,12]
y = [2,4,6,8]
plt.plot(x, y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

#### 2.Plot multiple lines on the same graph.

```python
x2 = [2, 4, 6, 8, 10]
y2 = [1, 2, 6, 9, 14]
y3 = [3, 6, 9, 12, 15]
y4 = [1, 3, 5, 7, 9]

plt.plot(x2, y2, label='1')
plt.plot(x2, y3, label='2')
plt.plot(x2, y4, label='3')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

#### 3. Add markers to a line plot.

```python
plt.plot(x, y, marker='o', markersize=10, color='red', linewidth=2, label='Line with Markers')
plt.show()
```

#### 4. Customize line styles and colors.

```python
plt.plot(x2, y2, linestyle='-', color='red', linewidth=2, label='Solid Red Line')
plt.plot(x2, y3, linestyle='--', color='green', linewidth=3, label='Dashed Green Line')
plt.plot(x2, y4, linestyle=':', color='blue', linewidth=1, label='Dotted Blue Line')
```

##  Scatter Plots:

#### 1. Create a basic scatter plot.

```python
x = np.random.normal(0, 1, 30)
y = np.random.normal(0, 1, 30)

plt.scatter(x, y, s=50, c='red', alpha=1.0)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

#### 2. Customize marker styles, sizes, and colors.

```python
plt.scatter(x, y, s=100, c='red', marker='s', edgecolors='black', linewidths=2)
```

#### 3. Plot multiple scatter plots on the same graph.

```python
x1 = np.random.normal(0, 1, 50)
y1 = np.random.normal(0, 1, 50)
x2 = np.random.normal(2, 1, 50)
y2 = np.random.normal(2, 1, 50)
x3 = np.random.normal(-2, 1, 50)
y3 = np.random.normal(-2, 1, 50)

fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(x1, y1, s=50, c='blue', alpha=0.7, label='Set 1')
ax.scatter(x2, y2, s=50, c='red', alpha=0.7, label='Set 2')
ax.scatter(x3, y3, s=50, c='green', alpha=0.7, label='Set 3')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
```

#### 4. Add labels and titles to the scatter plot.

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.scatter(x, y, s=50, c='blue', alpha=1.0)

ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
```

## Bar Charts:

#### 1. Create a vertical bar chart.

```python
labels = ['A', 'B', 'C', 'D', 'E']
values = [50, 55, 60, 30, 20]

fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(labels, values, width=0.6, color='blue')

ax.set_xlabel('Atrributes')
ax.set_ylabel('Threshold')

plt.xticks(rotation=45)
```

#### 2.Create a horizontal bar chart.


```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.barh(labels, values, height=0.6, color='blue')

ax.set_xlabel('Threshold')
ax.set_ylabel('Attributes')
```

#### 3. Customize bar colors and add edge colors.

```python
b_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
e_colors = ['#0b547a', '#b35900', '#1a6519', '#8c1d20', '#623c7a']

fig, ax = plt.subplots(figsize=(8, 6))

ax.bar(labels, values, width=0.6, color=b_colors, edgecolor=e_colors, linewidth=2)

ax.set_xlabel('Attributes')
ax.set_ylabel('Threshold')

ax.grid(axis='x', color='lightgray', linestyle='-', linewidth=1)

plt.show()
```

# Month 01 (Mini-project) Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

import_dataFrame = pd.read_csv('uni_data.csv')
```

# Data Exploration and Cleaning


```python
print(import_dataFrame.head())
print(import_dataFrame.info())
print(import_dataFrame.describe())
```

# Handle missing values & Remove Missing Values

```python
null_val = import_dataFrame.isnull()
print(null_val)
null_drop = import_dataFrame.dropna()
print(null_drop)
null_drop = import_dataFrame.dropna(axis=1)
print(null_drop)
print(import_dataFrame)
import_dataFrame['pre_requirement'] = import_dataFrame['pre_requirement'].fillna('Beginner')
print(import_dataFrame)
```

# Handling the Data Types

```python
print(import_dataFrame.dtypes)

import_dataFrame = import_dataFrame.astype({'course_type': 'string', 'pre_requirement': 'string', 'department': 'string'})
import_dataFrame['pre_score'] = pd.to_numeric(import_dataFrame['pre_score'], errors='coerce')

print(import_dataFrame.dtypes)
```

# Graphical Computations

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='course_type', y='enrollment_count', data=import_dataFrame)
plt.title('Enrollment Count by Course Type')
plt.xlabel('Course Type')
plt.ylabel('Enrollment Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x='pre_score', y='post_score', data=import_dataFrame, color='red', marker='o', label='Post-score')
plt.scatter(x='pre_score', y='pre_score', data=import_dataFrame, color='blue', marker='o', label='Pre-score')
plt.title('Pre-score vs Post-score')
plt.xlabel('Pre-score')
plt.ylabel('Post-score')
plt.show()
```

# Statistical Analysis

```python
online_enrollment = import_dataFrame[import_dataFrame['course_type'] == 'online']['enrollment_count']
classroom_enrollment = import_dataFrame[import_dataFrame['course_type'] == 'classroom']['enrollment_count']

t_stat, p_value = stats.ttest_ind(online_enrollment, classroom_enrollment)

print(f"t-statistic: {t_stat:.2f}")
print(f"p-value: {p_value:.4f}")
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='enrollment_count', hue='course_type', data=import_dataFrame)
plt.title('Enrollment Trend by Course Type')
plt.xlabel('Year')
plt.ylabel('Enrollment Count')
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='pre_requirement', y='enrollment_count', data=import_dataFrame)
plt.title('Enrollment by Pre-requirement')
plt.xlabel('Pre-requirement')
plt.ylabel('Enrollment Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
import_dataFrame['enrollment_count'].hist()
plt.title('Enrollment Count Distribution')

plt.subplot(1, 2, 2)
import_dataFrame['pre_score'].hist()
plt.title('Pre-Score Distribution')
plt.show()
```

## Subplots:

#### 1. Generate a figure with 2x2 subplots, each containing a different type of plot (e.g., line plot, scatter plot, bar plot, and histogram).

```python
import numpy as np
import matplotlib.pyplot as mtp
```

```python
rand_1 = np.random.randint(0, 40, 50)
rand1_sin_val = np.sin(rand_1)
rand_norm = np.random.normal(0, 1, 50)
rand_poisson = np.random.poisson(5, 50)
random_norm2 = np.random.normal(loc=0, scale=1, size=10)

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].plot(rand_1, rand1_sin_val)
axs[0, 0].set_title('Line')

axs[0, 1].scatter(rand_1, rand_norm)
axs[0, 1].set_title('Scatter')

axs[1, 0].bar(rand_1, rand_poisson)
axs[1, 0].set_title('Bar')

axs[1, 1].hist(random_norm2, bins=20)
axs[1, 1].set_title('Histogram')
```

#### 2. Customize the subplots with titles, axis labels, and legends.

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0, 0].plot(rand_1, rand1_sin_val, color='blue', linewidth=2)
axs[0, 0].set_title('Line Plot', fontsize=14, fontweight='bold')
axs[0, 0].set_xlabel('X', fontsize=12)
axs[0, 0].set_ylabel('Y', fontsize=12)

axs[0, 1].scatter(rand_1, rand_norm, color='red', s=50)
axs[0, 1].set_title('Scatter Plot', fontsize=14, fontweight='bold')
axs[0, 1].set_xlabel('X', fontsize=12)
axs[0, 1].set_ylabel('Y', fontsize=12)

axs[1, 0].bar(rand_1, rand_poisson, color='green', width=0.6)
axs[1, 0].set_title('Bar Plot', fontsize=14, fontweight='bold')
axs[1, 0].set_xlabel('X', fontsize=12)
axs[1, 0].set_ylabel('Y', fontsize=12)

axs[1, 1].hist(random_norm2, bins=20, color='orange')
axs[1, 1].set_title('Histogram', fontsize=14, fontweight='bold')
axs[1, 1].set_xlabel('Value', fontsize=12)
axs[1, 1].set_ylabel('Frequency', fontsize=12)
```

## Histograms:

#### 1. Create a histogram for a given numerical data set. Customize the number of bins.

```python
num_bins = 2
fig, ax = plt.subplots(figsize=(8, 6))
ax.hist(rand_norm, bins=num_bins, color='blue', edgecolor='black')

ax.set_title('Histogram of Random Data', fontsize=16, fontweight='bold')
ax.set_xlabel('Value', fontsize=14)
ax.set_ylabel('Frequency', fontsize=14)
```

#### 2. Plot multiple histograms on the same figure to compare different data sets.

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(rand_norm, bins=20, alpha=0.5, color='blue')
ax.hist(random_norm2, bins=20, alpha=0.5, color='orange')
ax.hist(rand_poisson, bins=20, alpha=0.5, color='green')

```

#### 3. Normalize the histograms and add appropriate labels and titles.

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(rand_norm, bins=20, alpha=0.5, color='blue', label='Sales')
ax.hist(random_norm2, bins=20, alpha=0.5, color='orange', label='Marketing')
ax.hist(rand_poisson, bins=20, alpha=0.5, color='green', label='Accounts')

ax.set_title('Comparison', fontsize=16, fontweight='bold')
ax.set_xlabel('Days', fontsize=14)
ax.set_ylabel('Hours', fontsize=14)

ax.legend(loc='upper right', fontsize=12)
```

## Box Plots:

#### 1. Create a box plot for a given numerical data set.

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.boxplot(rand_norm, patch_artist=True, boxprops=dict(facecolor='lightblue'))

ax.set_title('Numerical Data', fontsize=16, fontweight='bold')
ax.set_xlabel('Data', fontsize=14)
ax.set_ylabel('Value', fontsize=14)
```

#### 2. Generate multiple box plots in a single figure to compare different data sets.

```python
fig, ax = plt.subplots(figsize=(10, 6))

box_props = dict(linestyle='-', linewidth=2, color='black')
ax.boxplot([rand_norm, random_norm2, rand_poisson], patch_artist=True, boxprops=box_props,)

ax.set_xticklabels(['1', '2', '3'])

ax.set_title('Box Plots', fontsize=16, fontweight='bold')
ax.set_xlabel('Data Set', fontsize=14)
ax.set_ylabel('Value', fontsize=14)
```

#### 3. Customize the box plots with titles, labels, and change the colors

```python
fig, ax = plt.subplots(figsize=(10, 6))

box_props = dict(linestyle='-', linewidth=2, color='black')
ax.boxplot([rand_norm, random_norm2, rand_poisson], patch_artist=True, boxprops=box_props)

ax.set_xticklabels(['1', '2', '3'])

ax.set_title('Box Plots', fontsize=16, fontweight='bold')
ax.set_xlabel('Frequency', fontsize=14)
ax.set_ylabel('Threshold', fontsize=14)
```

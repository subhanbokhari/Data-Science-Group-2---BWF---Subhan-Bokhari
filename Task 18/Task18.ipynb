## 1. Scatter Plot

```python
import pandas as pd
import numpy as np
import seaborn as sns

import_dataFrame = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
import_dataFrame2 = pd.DataFrame(np.random.randint(0, 100, size=(15, 3)), columns=['X', 'Y', 'Z'])
import_dataFrame3 = pd.DataFrame(np.random.normal(0, 1, size=(20, 4)), columns=['P', 'Q', 'R', 'S'])
import_dataFrame4 = pd.DataFrame(np.random.choice(['apple', 'banana', 'cherry', 'grapefruit'], size=(12, 2)), columns=['Fruit1', 'Fruit2'])
```

```python
sns.relplot(x="A", y="B", hue="D", size="E",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=import_dataFrame)
```

```python
sns.set_theme(style="whitegrid", palette="muted")
ax = sns.swarmplot(data=import_dataFrame2, x="X", y="Y", hue="Z")
ax.set(ylabel="")
```

```python
sns.set_theme(style="ticks")
sns.pairplot(import_dataFrame3, hue="P")
```

```python
palette = sns.cubehelix_palette(rot=-.2, n_colors=4)
g = sns.relplot(
    data=import_dataFrame4,
    x="Fruit1", y="Fruit2",
    hue="Fruit1", size="Fruit2",
    palette=palette, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```

## 2. Line Plot

```python
sns.catplot(
    data=import_dataFrame, x="A", y="B", hue="C",
    native_scale=True, zorder=1
)
sns.regplot(
    data=import_dataFrame, x="D", y="E",
    scatter=False, truncate=False, order=2, color=".2",
)
```

```python
sns.lineplot(data=import_dataFrame2, palette="tab10", linewidth=2.5)
```

```python
dots = sns.load_dataset("dots")

palette = sns.color_palette("rocket_r")

sns.relplot(
    data=dots,
    x="time", y="firing_rate",
    hue="coherence", size="choice", col="align",
    kind="line", size_order=["T1", "T2"], palette=palette,
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```

## 3. Bar Plot

```python
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data=import_dataFrame, kind="bar",
    x="A", y="B", hue="C",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Body mass (g)")
g.legend.set_title("")
```

## 4. Box Plot

```python
sns.set_theme(style="ticks", palette="pastel")

sns.boxplot(x="A", y="B",
            hue="C", palette="pastel",
            data=import_dataFrame)

sns.despine(offset=10, trim=True)
```

```python
sns.set_theme(style="ticks")

ax.set_xscale("log")

sns.boxplot(
    data=import_dataFrame, x="A", y="B", hue="C",
    whis=[0, 100], width=.6, palette="vlag"
)
sns.stripplot(data=import_dataFrame, x="D", y="E", size=4, color=".3")

ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

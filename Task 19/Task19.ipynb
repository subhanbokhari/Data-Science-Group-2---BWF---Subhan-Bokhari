## Sample Data Sets

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data1 = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 28],
    'Kids': [5, 2, 1],
    'City': ['New York', 'Chicago', 'Los Angeles']
}
df1 = pd.DataFrame(data1)

data2 = {
    'Product': ['Laptop', 'Smartphone', 'Tablet'],
    'Price': [999.2, 599.7, 399.8],
    'Quantity': [10, 20, 15]
}
df2 = pd.DataFrame(data2)

data3 = np.random.rand(10, 10)
df3 = pd.DataFrame(data3, columns=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'])

print("Dataframe 1:")
print(df1)
print("\nDataframe 2:")
print(df2)
print("\nDataframe 3:")
print(df3)
```

## 1.Linear Regression Plot (sns.regplot)

```python
sns.regplot(x='Age', y='Kids', data=df1)
plt.xlabel('Age')
plt.ylabel('Kids')
plt.show()
```

## 2. Multiple Linear Regression Plot (sns.lmplot)

```python
sns.lmplot(x='Price', y='Quantity', data=df2, hue='Product', height=6, aspect=1.5)
plt.xlabel('Price')
plt.ylabel('Quantity')
```

## 3. Residual Plot (sns.residplot)

```python
sns.residplot(x='Age', y='Age', data=df1)
plt.xlabel('Age')
plt.ylabel('Residuals')
plt.show()
```

## 4. Histogram (sns.histplot)

```python
sns.histplot(data=df2, x='Price', hue='Product', multiple='stack', bins=10)
plt.xlabel('Price')
plt.ylabel('Count')
```

## 5. Kernel Density Estimate (KDE) Plot (sns.kdeplot)

```python
sns.kdeplot(data=df1, x='Age', hue='City', fill=True, warn_singular=False)
plt.xlabel('Age')
plt.ylabel('Density')
```

## 6. Distribution Plot (sns.displot)

```python
sns.displot(data=df2, x='Price', kde=True, bins=5, color='blue')
plt.xlabel('Price')
plt.ylabel('Density')
```

## 7. Box Plot (sns.boxplot)

```python
sns.boxplot(data=df1, x='Name', y='Kids', color='lightblue')
plt.xlabel('Name')
plt.ylabel('Kids')
```

## 8. Violin Plot (sns.violinplot)

```python
sns.violinplot(data=df2, x='Product', y='Price', inner='quartile', hue='Product', legend=False)
plt.xlabel('Product')
plt.ylabel('Price')
```

## 9.Pair Plot (sns.pairplot)

```python
sns.pairplot(data=df1, vars=['Age', 'Kids'])
```

## 10. Heatmap (sns.heatmap)

```python
sns.heatmap(df3.corr(), annot=True, cmap='YlOrRd')
plt.title('Heatmap of the Random DataFrame')
plt.show()
```

## 11. Joint Plot (sns.jointplot)

```python
sns.jointplot(x='A', y='B', data=df3, kind='scatter')
```

## 13. Facet Grid (sns.FacetGrid)

```python
data4 = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
    'Sales': [100, 80, 120, 90, 110],
    'Profit': [20, 15, 25, 18, 22],
    'Inventory': [50, 40, 60, 45, 55],
    'Marketing_Spend': [10, 8, 12, 9, 11]
}

df4 = pd.DataFrame(data4)

g = sns.FacetGrid(df4, row='Inventory', col='Marketing_Spend', hue='Sales')
g.map(sns.scatterplot, 'Inventory', 'Profit')
g.add_legend()
```
